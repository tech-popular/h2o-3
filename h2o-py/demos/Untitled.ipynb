```python
# run h2o Kmeans

# Import h2o library
import h2o
from h2o.estimators import H2OKMeansEstimator

# init h2o cluster
h2o.init(strict_version_check=False)
```

```python
# load data
import pandas as pd

data = pd.read_csv("../../smalldata/chicago/chicagoAllWeather.csv")
data = data.iloc[:,[1, 2, 3, 4, 5]]
print(data.shape)
data.head()
```

```python
# import time to measure elapsed time
from timeit import default_timer as timer
from datetime import timedelta
import time

start = timer()
end = timer()
print("Time:", timedelta(seconds=end-start))
```

```python
data_h2o = h2o.H2OFrame(data)

# run h2o Kmeans
h2o_km_co_t = H2OKMeansEstimator(k=3, init="furthest", cluster_size_constraints=[1000, 2000, 1000], standardize=True)
start = timer()
h2o_km_co_t.train(training_frame=data_h2o)
end = timer()

# show details
h2o_km_co_t.show()
time_km_co_5000 = timedelta(seconds=end-start)
print("Time:", time_km_co_5000)
```
